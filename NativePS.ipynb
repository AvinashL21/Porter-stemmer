```python
file=open("/content/sample_data/Football.txt")
text=file.readline()
file.close()
print(text)
```

```python
len(text)
```

```python
def number_of_words(text):
    count = 0
    for line in text.split('\n'):
        count += len(line.split())
    return count
Original_no_of_words=number_of_words(text)
print(Original_no_of_words)
```

```python
 import nltk
 nltk.download('punkt')
 from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
tokenized_words=word_tokenize(text)
```

```python
tokenized_words
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))
```

```python
filtered_list = []
for word in tokenized_words:
   if word.casefold() not in stop_words:
        filtered_list.append(word)
filtered_list
```

```python
filtered_list = [
    word for word in tokenized_words if word.casefold() not in stop_words
]
filtered_list
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
stemmed_words = [stemmer.stem(word) for word in filtered_list]
```

```python
print(stemmed_words)
```

```python
Number_of_stemmed_words=number_of_words(' '.join(stemmed_words))
Number_of_stemmed_words
```

```python
#Rate of Reduction of words
Reduction_rate=((Number_of_stemmed_words/Original_no_of_words)*100)
Reduction_rate
```
